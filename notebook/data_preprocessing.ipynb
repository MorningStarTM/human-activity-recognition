```python
def sample_dict(df1, df2, path):
    d = {'acc_x_mean':df1['x'].mean(),
        'acc_y_mean':df1['y'].mean(),
        'acc_z_mean':df1['z'].mean(),
        'acc_x_std':df1['x'].std(),
        'acc_y_std':df1['y'].std(),
        'acc_z_std':df1['z'].std(),
        'acc_x_max':df1['x'].max(),
        'acc_y_max':df1['y'].max(),
        'acc_z_max':df1['z'].max(),
        'acc_x_min':df1['x'].min(),
        'acc_y_min':df1['y'].min(),
        'acc_z_min':df1['z'].min(),
        'gyro_x_mean':df2['x'].mean(),
        'gyro_y_mean':df2['y'].mean(),
        'gyro_z_mean':df2['z'].mean(),
        'gyro_x_std':df2['x'].std(),
        'gyro_y_std':df2['y'].std(),
        'gyro_z_std':df2['z'].std(),
        'gyro_x_max':df2['x'].max(),
        'gyro_y_max':df2['y'].max(),
        'gyro_z_max':df2['z'].max(),
        'gyro_x_min':df2['x'].min(),
        'gyro_y_min':df2['y'].min(),
        'gyro_z_min':df2['z'].min(),
        'activity':path.split('\\')[-2]
                   }
    return d
    
```

```python
def read_data(df, path):
    l = glob(os.path.join(path, "*", "*"))
    for i in range(0,len(l),2):
        print(l[i])
        d1, d2 = pd.read_csv(l[i]), pd.read_csv(l[i+1])
        dic = sample_dict(d1, d2, l[i])
        print(dic['activity'])
        df = df.append(dic, ignore_index=True)
    return df
```
